```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data = pd.read_csv("dataset/order_data.csv", delimiter=" ", header=None)
data.head()
```

Support: C'est la mesure de la popularité d'un ensemble d'éléments, c'est-à-dire le nombre de transactions dans lesquelles un élément apparaît par rapport au nombre total de transactions. Il s’agit simplement de la probabilité qu’un client achète un article. La formule mathématique pour représenter le support de l'élément X est :

### Support(X) = (Transactions containing (X))/(Total Transactions)

__________________________________________________________________________________________________________________________________

Confiance: Ce paramètre détermine l&#39;impact d&#39;un produit sur un autre, c&#39;est-à-dire la probabilité que si une personne achète le produit X, elle achètera également le produit Y. Sa représentation en termes mathématiques est:

### Confiance({X} =&gt; {Y}) = (Transactions contenant à la fois X et Y)/(Transactions contenant X)

__________________________________________________________________________________________________________________________________

Lift: Surmonte les limites de la mesure de la confiance et calculera une nouvelle confiance qui tient compte de la popularité des deux éléments. La représentation mathématique du Lift est :

### Lift({X} =&gt; {Y}) = Confience({X} =&gt; {Y}) / Support(Y)

Si la mesure Lift est supérieure à 1, cela signifie que Y est susceptible d'être acheté avec X, tandis qu'une valeur inférieure à 1 indique qu'il est peu probable que Y soit acheté avec X. Une valeur de Lift proche de 1 indique que les deux ensembles d'éléments dans les transactions apparaissent souvent ensemble mais il n'y a aucune association entre elles.

__________________________________________________________________________________________________________________________________

```python
data_list = []
for row in range(0, 20):
    data_list.append([str(data.values[row, col]) for col in range(0, 9)])

data_list
```

```python
association_rules = apriori(data_list, min_support=0.25, min_confidence=0.2, min_lift=2, min_length=2)
results = list(association_rules)
```

```python
for i in range(0, len(results)):
    print(f"Required Association No. {i+1} is: {results[i]}")
    print('-'*25)
```
